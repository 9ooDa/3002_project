```python
from sklearn.preprocessing import LabelEncoder

df_new = df

le = LabelEncoder()

df_new['Age'] = le.fit_transform(df['Age'])
```

```python
df_new.loc[:,'Segmentation'] = df_new['Segmentation'].replace(['A', 'B', 'C', 'D'], [1, 2, 3, 4])

import random

for i, row in df_new.iterrows():
    sp_val = 0
    if row['Spending_Score'] == 'Low':
        sp_val = random.randrange(0, 34)
        
    elif row['Spending_Score'] == 'Average':
        sp_val = random.randrange(34, 67)
        
    elif row['Spending_Score'] == 'High':
        sp_val = random.randrange(67, 100)

    df_new.at[i,'Spending_Score'] = sp_val
    
data = df_new[['Age','Spending_Score']]

df_new = df_new.dropna()
data
```
